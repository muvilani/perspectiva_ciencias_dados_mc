# Comandos Básicos de Git, Terminal e Ferramentas

Este notebook reúne todos os comandos essenciais vistos no curso até agora para configuração de ambiente, versionamento de código com Git, gerenciamento de pacotes Python com UV e edição de texto com Vim.

---

## 1. Configurando o GitHub (Token de Acesso)

Para que sua máquina consiga se comunicar com o GitHub (fazer o push do seu código), é necessário usar um token de autenticação em vez de senha.

**Passo a passo para criar o Token Classic:**
1. No GitHub, vá em **Settings** (Configurações).
2. No menu lateral esquerdo, desça até o final e clique em **Developer settings**.
3. Clique em **Personal access tokens** -> **Tokens (classic)**.
4. Clique no botão **Generate new token** -> **Generate new token (classic)**.
5. Dê um nome, marque as caixinhas de permissão necessárias (normalmente marcamos a caixa principal `repo`) e gere o token.
6. Copie o token (ele começa com `ghp_...`) e **nunca** o compartilhe publicamente.

## 2. Comandos Básicos de Terminal

Comandos essenciais para navegar pelo computador sem usar o mouse (válidos para Linux, Mac e Git Bash no Windows).

* **`ls`**: Lista os arquivos e pastas do diretório atual.
  * *Exemplo:* `ls` (mostra o que tem na pasta) ou `ls -a` (mostra até os arquivos ocultos).
* **`cd`**: Acessa uma pasta específica.
  * *Exemplo:* `cd Projetos` (entra na pasta "Projetos").
* **`cd ..`**: Volta uma pasta para trás (sobe um nível).
  * *Exemplo:* `cd ..`
* **`cat`**: Lê e imprime o conteúdo de um arquivo no terminal.
  * *Exemplo:* `cat arquivo.txt`
* **`touch`**: Cria um arquivo vazio na pasta atual.
  * *Exemplo:* `touch .gitignore` (cria o arquivo que diz ao git o que ignorar).

**Atalhos Úteis:**
* **`Ctrl + Alt + T`**: Abre o terminal (no Linux). No Windows, clique com o botão direito na pasta e vá em "Open Git Bash here".
* **`Ctrl + L`**: Limpa a tela do terminal.
* **`Ctrl + H`**: Mostra os arquivos ocultos (no Linux). No Windows, você vai em Exibir -> Mostrar -> Itens ocultos.

## 3. Comandos Básicos do Git

O Git é o sistema que gerencia as versões do seu código, tirando "fotos" do projeto ao longo do tempo.

* **`git clone`**: Baixa um repositório inteiro do GitHub para a sua máquina.
  * *Exemplo:* `git clone https://github.com/usuario/meu-projeto.git`
* **`git status`**: Mostra o status atual do repositório. Mostra o que foi alterado (em vermelho) e o que está pronto para ser salvo (em verde).
  * *Exemplo:* `git status`
* **`git diff`**: Mostra as diferenças exatas, linha por linha, do que foi alterado no código antes de você adicionar.
  * *Exemplo:* `git diff`
* **`git add`**: Prepara os arquivos modificados para serem salvos (coloca no "palco" ou *staging area*).
  * *Exemplo 1:* `git add f1` (adiciona apenas o arquivo f1).
  * *Exemplo 2:* `git add .` (adiciona TODOS os arquivos e pastas alterados).
* **`git restore --staged`**: Remove um arquivo que você preparou com o `git add` caso tenha adicionado por engano (tira do "palco").
  * *Exemplo:* `git restore --staged f1`
* **`git commit`**: Salva a versão atual do projeto (tira a "foto"). Sempre exige uma mensagem explicando o que foi feito.
  * *Exemplo:* `git commit -m "Adiciona notebook com os comandos do curso"`
* **`git pull`**: Puxa as novidades do GitHub para o seu computador. Sempre faça isso antes de enviar seu código para evitar conflitos.
  * *Exemplo:* `git pull`
* **`git push`**: Envia as suas atualizações salvas (commits) do computador para o GitHub.
  * *Exemplo:* `git push`
* **`git log`**: Mostra o histórico de commits do projeto, quem fez e quando. (Dica: aperte a letra "Q" para sair da lista se ela for muito longa).
  * *Exemplo:* `git log`

## 4. UV - Gerenciador de Pacotes Python

O `uv` é uma ferramenta super rápida para gerenciar projetos e dependências em Python.

* **`uv init`**: Inicia um projeto novo do zero na pasta atual.
  * *Exemplo:* `uv init`
* **`uv add`**: Instala/baixa pacotes externos para o seu projeto.
  * *Exemplo:* `uv add pandas` ou `uv add requests`
* **`uv run`**: Executa um arquivo ou pacote Python usando o ambiente criado.
  * *Exemplo:* `uv run script.py`

## 5. Edição de Texto no Terminal com Vim / Vi

O Vim (ou vi) é um editor de texto que roda direto no terminal. Ele possui modos diferentes (um para digitar e outro para comandos).

* **Abrir o editor:**
  * *Exemplo:* `vi meuarquivo.txt` (Abre o arquivo. Se não existir, ele cria).
* **`i` (Insert):** Entra no modo de inserção. Aperte `i` quando quiser começar a digitar o texto.
* **`ESC`:** Sai do modo de inserção e volta para o modo de comandos. (Aperte sempre antes de tentar salvar ou sair).
* **`o`**: Pula uma linha para baixo e já entra automaticamente no modo de inserção.
* **`u` (Undo):** Desfaz a última alteração (aperte no modo de comandos).
* **`:w` (Write):** Salva o arquivo. (Digite `ESC` depois `:w` e `Enter`).
* **`:q!` (Quit):** Sai do Vim forçadamente sem salvar.
