# Pegando os dados da notícia dos carros mais vendidos de Março de 2026

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "https://motor1.uol.com.br/news/791853/carros-mais-vendidos-marco-onix/"

def extrair_tabela_vendas(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    
    print(f"Acessando: {url}")
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Encontrar a tabela no conteúdo do artigo
    table = soup.find('table') # <--- Aqui procuramos a tag <table>
    if not table:
        print("Tabela não encontrada.")
        return None
    
    rows = table.find_all('tr') # <--- Dentro da tabela, procuramos todas as tags <tr> 
    data = []
    
    for row in rows:
        cols = row.find_all('td') # <--- Para cada linha, procuramos todas as tags <td> (células)
        cols = [ele.text.strip() for ele in cols] # <--- Extraímos o texto de cada célula e removemos espaços extras
        if cols:
            data.append(cols) # <--- Adicionamos a lista de células (uma linha) à nossa lista de dados
    
    # A primeira linha costuma ser o cabeçalho
    if data:
        # Criamos um DataFrame do Pandas. A primeira linha de 'data' são os cabeçalhos.
        df = pd.DataFrame(data[1:], columns=data[0]) 
        return df
    return None

if __name__ == "__main__":
    url = "https://motor1.uol.com.br/news/791853/carros-mais-vendidos-marco-onix/"
    df_vendas = extrair_tabela_vendas(url )
    
    if df_vendas is not None:
        col_vendas = df_vendas.columns[2]
        df_vendas[col_vendas] = df_vendas[col_vendas].str.replace('.', '', regex=False).astype(int)
        
        print("Dados extraídos com sucesso!")
        print(df_vendas)
```

# Pegando dados da notícia dos carros mais baratos de 2026

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

def extrair_precos_v4(url):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    
    print(f"Acessando: {url}")
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Encontramos todos os itens de lista (<li>)
    list_items = soup.find_all('li')
    
    precos_data = []
    current_car = {}
    
    for li in list_items:
        # .get_text() com strip=True remove espaços extras e lida com &nbsp;
        text = li.get_text(strip=True)
        
        # Normalizamos o texto: removemos espaços duplos e caracteres invisíveis
        # Isso transforma "Preço:  R$ 78.690" em "Preço: R$ 78.690"
        text_norm = " ".join(text.split())
        
        # Procuramos por "Versão:" de forma insensível a maiúsculas/minúsculas
        if re.search(r'Versão:', text_norm, re.IGNORECASE):
            # Se já tínhamos um carro sendo processado, salvamos antes de começar o próximo
            if current_car and 'preco' in current_car:
                precos_data.append(current_car)
            
            # Extraímos o nome do modelo removendo a palavra "Versão:"
            modelo_completo = re.sub(r'Versão:\s*', '', text_norm, flags=re.IGNORECASE).strip()
            # Criamos uma chave simplificada para cruzamento (ex: "RENAULT KWID")
            marca_modelo = " ".join(modelo_completo.split()[:2]).upper()
            current_car = {'modelo_completo': modelo_completo, 'modelo_chave': marca_modelo}
            
        # Procuramos por "Preço:" de forma insensível a maiúsculas/minúsculas
        elif re.search(r'Preço:', text_norm, re.IGNORECASE) and current_car:
            # Usamos Regex para capturar apenas os números após o "R$"
            # \d+ captura sequências de dígitos
            match = re.search(r'R\$\s*([\d\.]+)', text_norm)
            if match:
                # Removemos o ponto de milhar e convertemos para número inteiro
                valor_limpo = match.group(1).replace(".", "")
                current_car['preco'] = int(valor_limpo)
                
    # Não esquecer de adicionar o último carro da lista
    if current_car and 'preco' in current_car:
        precos_data.append(current_car)
    
    return pd.DataFrame(precos_data)

if __name__ == "__main__":
    url = "https://motor1.uol.com.br/features/783284/carros-mais-baratos-brasil-2026/"
    df_precos = extrair_precos_v4(url)
    
    if not df_precos.empty:
        # Limpeza final: removemos duplicatas e garantimos que o preço existe
        df_precos = df_precos.drop_duplicates(subset=['modelo_chave']).dropna(subset=['preco'])
        # Removemos itens que não são carros (como erros de extração)
        df_precos = df_precos[df_precos['modelo_chave'] != 'TECHNO']
        
        print("\n--- DADOS EXTRAÍDOS COM SUCESSO ---")
        print(df_precos[['modelo_chave', 'preco']])
    else:
        print("\nERRO: Nenhum dado foi encontrado. Verifique a estrutura da página.")
```

# Pegando dados da notícia dos preços do primeiro trimestre

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extrair_vendas_motor1_puro(url_alvo):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    
    print(f"Acessando: {url_alvo}")
    response = requests.get(url_alvo, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Busca a primeira tabela no HTML
    tabela = soup.find('table')
    
    if not tabela:
        print("Nenhuma tabela encontrada.")
        return None

    dados = []
    for linha in tabela.find_all('tr'):
        colunas = [col.get_text().strip() for col in linha.find_all(['td', 'th'])]
        if colunas:
            dados.append(colunas)

    # Cria o DataFrame a partir da lista de listas
    if dados:
        # A primeira linha costuma ser o cabeçalho
        df = pd.DataFrame(dados[1:], columns=dados[0])
        return df
    return None

# URL do 1º Trimestre
url = "https://motor1.uol.com.br/news/792306/venda-carros-novos-1trimetre-2026/"
df_resultado = extrair_vendas_motor1_puro(url)

if df_resultado is not None:
    print("\n--- DADOS EXTRAÍDOS (SEM DEPENDER DE LXML) ---")
    print(df_resultado)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

def extrair_tabelas_motor1(url_alvo):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"
    }
    
    print(f"Acessando: {url_alvo}")
    try:
        response = requests.get(url_alvo, headers=headers)
        response.raise_for_status()
        soup = BeautifulSoup(response.text, 'html.parser')

        todas_tabelas_html = soup.find_all('table')
        
        if not todas_tabelas_html:
            print("Nenhuma tabela encontrada no site.")
            return []

        lista_dataframes = []

        for i, tabela_html in enumerate(todas_tabelas_html):
            linhas_extraidas = []
            for tr in tabela_html.find_all('tr'):
                # Extrai o texto de cada célula, removendo espaços e quebras de linha
                colunas = [td.get_text(strip=True) for td in tr.find_all(['td', 'th'])]
                if colunas:
                    linhas_extraidas.append(colunas)

            if len(linhas_extraidas) > 1:
                # Cria o DataFrame usando a primeira linha como cabeçalho
                df = pd.DataFrame(linhas_extraidas[1:], columns=linhas_extraidas[0])
                lista_dataframes.append(df)
                print(f"Tabela {i+1} extraída com sucesso ({len(df)} linhas).")

        return lista_dataframes

    except Exception as e:
        print(f"Erro na extração: {e}")
        return []

# URL da notícia
url = "https://motor1.uol.com.br/news/792306/venda-carros-novos-1trimetre-2026/"
tabelas = extrair_tabelas_motor1(url)

if tabelas:
    # Tabela 1 costuma ser o ranking de Marcas
    print("\n--- TABELA 1: RANKING DE MARCAS ---")
    print(tabelas[0].head)

    # Tabela 2 costuma ser o ranking de Modelos
    if len(tabelas) > 1:
        print("\n--- TABELA 2: RANKING DE MODELOS (TOP 50) ---")
        # Ajustando nomes de colunas se necessário (opcional)
        df_modelos = tabelas[1]
        print(df_modelos.head)
        
```

# Organizando os dados 

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import os

OUTPUT_DIR = "dados_carros"
os.makedirs(OUTPUT_DIR, exist_ok=True)

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0.0.0 Safari/537.36"
    )
}


def get_soup(url):
    r = requests.get(url, headers=HEADERS)
    r.raise_for_status()
    return BeautifulSoup(r.text, "html.parser")


def salvar(df, nome, descricao):
    caminho = os.path.join(OUTPUT_DIR, nome)
    df.to_csv(caminho, index=False, encoding="utf-8-sig")
    print(f"  ✅ {descricao} → {nome} ({len(df)} linhas)")


# ── 1. Vendas de março 2026 ──────────────────────────────
print("1. Extraindo vendas de março 2026...")
soup = get_soup("https://motor1.uol.com.br/news/791853/carros-mais-vendidos-marco-onix/")
tabela = soup.find("table")
dados = []
for tr in tabela.find_all("tr"):
    cols = [td.get_text(strip=True) for td in tr.find_all(["td", "th"])]
    if cols:
        dados.append(cols)

df_marco = pd.DataFrame(dados[1:], columns=dados[0])
col_vendas = df_marco.columns[2]
df_marco[col_vendas] = pd.to_numeric(
    df_marco[col_vendas].str.replace(".", "", regex=False), errors="coerce"
)
salvar(df_marco, "vendas_marco_2026.csv", "Vendas março 2026")


# ── 2. Carros mais baratos 2026 ──────────────────────────
print("2. Extraindo carros mais baratos 2026...")
soup = get_soup("https://motor1.uol.com.br/features/783284/carros-mais-baratos-brasil-2026/")
precos_data = []
current_car = {}

for li in soup.find_all("li"):
    text = " ".join(li.get_text(strip=True).split())
    if re.search(r"Versão:", text, re.IGNORECASE):
        if current_car and "preco" in current_car:
            precos_data.append(current_car)
        modelo_completo = re.sub(r"Versão:\s*", "", text, flags=re.IGNORECASE).strip()
        current_car = {
            "modelo_completo": modelo_completo,
            "modelo_chave": " ".join(modelo_completo.split()[:2]).upper(),
        }
    elif re.search(r"Preço:", text, re.IGNORECASE) and current_car:
        m = re.search(r"R\$\s*([\d\.]+)", text)
        if m:
            current_car["preco"] = int(m.group(1).replace(".", ""))

if current_car and "preco" in current_car:
    precos_data.append(current_car)

df_baratos = pd.DataFrame(precos_data)
df_baratos = (
    df_baratos
    .drop_duplicates(subset=["modelo_chave"])
    .dropna(subset=["preco"])
    .query("modelo_chave != 'TECHNO'")
    .reset_index(drop=True)
)
salvar(df_baratos, "carros_mais_baratos_2026.csv", "Carros mais baratos 2026")


# ── 3 e 4. Trimestre: marcas e modelos ──────────────────
print("3. Extraindo ranking do 1º trimestre 2026...")
soup = get_soup("https://motor1.uol.com.br/news/792306/venda-carros-novos-1trimetre-2026/")

todas_tabelas = []
for tabela_html in soup.find_all("table"):
    dados = []
    for tr in tabela_html.find_all("tr"):
        cols = [td.get_text(strip=True) for td in tr.find_all(["td", "th"])]
        if cols:
            dados.append(cols)
    if len(dados) > 1:
        # Corrige colunas duplicadas adicionando sufixo numérico
        cabecalho = dados[0]
        vistos = {}
        cabecalho_unico = []
        for col in cabecalho:
            if col in vistos:
                vistos[col] += 1
                cabecalho_unico.append(f"{col}_{vistos[col]}")
            else:
                vistos[col] = 0
                cabecalho_unico.append(col)

        df = pd.DataFrame(dados[1:], columns=cabecalho_unico)
        todas_tabelas.append(df)

# Tabela 0 = ranking de marcas
salvar(todas_tabelas[0], "trimestre_ranking_marcas.csv", "Ranking marcas 1º tri 2026")

# Tabelas 1, 2, 3 = modelos de jan, fev, mar
nomes_meses = ["jan_2026", "fev_2026", "mar_2026"]
dfs_modelos = []
for i, mes in enumerate(nomes_meses):
    idx = i + 1
    if idx < len(todas_tabelas):
        df_mes = todas_tabelas[idx].copy()
        df_mes["mes"] = mes
        salvar(df_mes, f"ranking_modelos_{mes}.csv", f"Ranking modelos {mes}")
        dfs_modelos.append(df_mes)

# Consolidado com os 3 meses
if dfs_modelos:
    df_consolidado = pd.concat(dfs_modelos, ignore_index=True)
    salvar(df_consolidado, "ranking_modelos_trimestre_completo.csv", "Consolidado trimestre")

print()
print(f"Todos os arquivos salvos em: ./{OUTPUT_DIR}/")
```
