Exercício 1

```python
import numpy as np

vetor = np.random.rand(20)
print(vetor)

media_vetor = np.mean(vetor)
print("Média do vetor:", media_vetor)
desvio_padrao_vetor = np.std(vetor)
print("Desvio padrão do vetor:", desvio_padrao_vetor)   
max_vetor = np.max(vetor)
print("Valor máximo do vetor:", max_vetor)
```

Exercício 2

```python
matriz_normal = np.random.randn(5, 5)
print("Essa é uma matriz 5x5 de números aleatórios seguindo uma distribuição normal (média 0 e desvio padrão 1):/n", matriz_normal)

soma_cada_linha = np.sum(matriz_normal, axis=1)
print("Soma de cada linha da matriz:", soma_cada_linha)

soma_cada_coluna = np.sum(matriz_normal, axis=0)
print("Soma de cada coluna da matriz:", soma_cada_coluna)

diag_principal = np.diag(matriz_normal)
print("Diagonal principal da matriz:", diag_principal)
```

Exercício 3

```python
vetor1 = np.array([1, 2, 3, 4, 5])
vetor2 = np.array([6, 7, 8, 9, 10])

prod_escalar = np.dot(vetor1, vetor2)
print("Produto escalar dos vetores:", prod_escalar)

dist_euclidiana = np.linalg.norm(vetor1 - vetor2)
print("Distância euclidiana entre os vetores:", dist_euclidiana)
```

Exercício 4

```python
matriz_inteiros = np.random.randint(-5, 6, (4, 4))
print("Matriz 4x4 de inteiros aleatórios entre -5 e 5:\n", matriz_inteiros)

invertivel = np.linalg.det(matriz_inteiros) 
print("Determinante da matriz de inteiros:\n", invertivel)
print("A matriz é invertível?" , "Sim" if invertivel != 0 else "Não")
inversao = np.linalg.inv(matriz_inteiros) 
print("Inversa da matriz de inteiros:", inversao)
identidade = np.dot(matriz_inteiros, inversao)
print("Verificando se o produto da matriz de inteiros pela sua inversa é a matriz identidade:\n", identidade)
```

Exercício 7

```python
media = [0, 0]
cov = [[1, 0], [0, 1]]
vetores_2d = np.random.multivariate_normal(media, cov, 100)

matriz_covariancia = np.cov(vetores_2d)
print("Matriz de covariância dos vetores 2D:\n", matriz_covariancia)
autovalores_cov, autovetores_cov = np.linalg.eig(matriz_covariancia)
print("Autovalores da matriz de covariância:", autovalores_cov) 
print("Autovetores da matriz de covariância:\n", autovetores_cov)

# Também nunca tinha visto isso, vou pesquisar e tirar dúvida sobre o código
direcao_maior_variancia = autovetores_cov[:, np.argmax(autovalores_cov)]
print("Direção de maior variância dos dados (autovetor correspondente ao maior autovalor):\n", direcao_maior_variancia)
```

Exercício 6

```python
medias_por_variedade = dados.groupby("variety")[["sepal.length", "sepal.width", "petal.length", "petal.width"]].mean()

variedades = medias_por_variedade.index
x = np.arange(len(variedades))
largura = 0.2

fig, ax = plt.subplots(figsize=(10, 6))

ax.bar(x + 0 * largura, medias_por_variedade["sepal.length"], width=largura, label="sepal.length")
ax.bar(x + 1 * largura, medias_por_variedade["sepal.width"],  width=largura, label="sepal.width")
ax.bar(x + 2 * largura, medias_por_variedade["petal.length"], width=largura, label="petal.length")
ax.bar(x + 3 * largura, medias_por_variedade["petal.width"],  width=largura, label="petal.width")

plt.title("Médias das medidas por variedade")
plt.xlabel("Variedade")
plt.ylabel("Valores médios")
plt.xticks(x + largura * 1.5, variedades, rotation=45)
plt.legend(title="Medidas")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

cores = {'Setosa': 'red', 'Versicolor': 'blue', 'Virginica': 'green'}

for variedade, cor in cores.items():
    subset = dados[dados["variety"] == variedade]
    ax.scatter(subset["sepal.length"], subset["sepal.width"], subset["petal.length"],
               c=cor, label=variedade, s=50, alpha=0.7)

ax.set_xlabel("Sepal Length")
ax.set_ylabel("Sepal Width")
ax.set_zlabel("Petal Length")
ax.set_title("Dispersão 3D por Variedade")
ax.legend(title="Variedade")
plt.tight_layout()
plt.show()
```
