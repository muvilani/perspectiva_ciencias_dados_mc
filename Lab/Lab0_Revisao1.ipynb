EXERCÍCIOS: Pandas, NumPy e Matplotlib

Passo 1 – Criar o dicionário principal

```python
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}
```

Passo 2 - Importar bibliotecas

```python
#!pip install numpy pandas matplotlib
import numpy as np # vetor
import pandas as pd # manipulação/visualização
import matplotlib.pyplot as plt # gráfico
```

Passo 3 - Gerar os dados 

```python
normal = np.random.normal(0, 1, size = 2)
exponencial = np.random.exponential(2, size = 2)
uniforme = np.random.uniform(-1, 1, size = 2)

#salvando em "dados"
distribuicoes["Normal"]["dados"] = normal
distribuicoes["Exponencial"]["dados"] = exponencial
distribuicoes["Uniforme"]["dados"] = uniforme
```

Passo 4 - Calculando métricas

```python
media_normal = np.mean(normal)
media_exponencial = np.mean(exponencial)
media_uniforme = np.mean(uniforme)
mediana_normal = np.median(normal)
mediana_exponencial = np
mediana_uniforme = np.median(uniforme)
var_normal = np.var(normal)
var_exponencial = np.var(exponencial)
var_uniforme = np.var(uniforme)
sd_normal = np.std(normal)
sd_exponencial = np.std(exponencial)
sd_uniforme = np.std(uniforme)

#Armazene em um dicionário dentro da chave "estatisticas"
distribuicoes["Normal"]["estatisticas"] = {"media": media_normal, "mediana": mediana_normal, "variancia": var_normal, "sd": sd_normal}
distribuicoes["Exponencial"]["estatisticas"] = {"media": media_exponencial, "mediana": mediana_exponencial, "variancia": var_exponencial, "sd": sd_normal}
distribuicoes["Uniforme"]["estatisticas"] = {"media": media_uniforme, "mediana": mediana_uniforme, "variancia": var_uniforme, "sd": sd_uniforme}
```

Passo 5 - Visualizar os dados

```python
hist_normal = plt.hist(normal)

hist_exponencial = plt.hist(exponencial)

hist_uniforme = plt.hist(uniforme)
```

Passo 6 - Organizar Resultados em Pandas

```python
df = pd.DataFrame(distribuicoes)
df
```

Passo 7 - Adicionar novas distribuições

```python
poisson = np.random.poisson(lam=2, size=2)

media_poisson = np.mean(poisson)
mediana_poisson = np.median(poisson)
var_poisson = np.var(poisson)
sd_poisson = np.std(poisson)

distribuicoes["Poisson"] = {
	"params": {"lam": 2},
	"dados": poisson,
	"estatisticas": {
		"media": media_poisson,
		"mediana": mediana_poisson,
		"variancia": var_poisson,
		"sd": sd_poisson
	}
}

df = pd.DataFrame(distribuicoes)
df
```

Passo 8 - Estatística adicionais

```python
min_normal = np.min(normal)
max_normal = np.max(normal)
min_exponencial = np.min(exponencial)
max_exponencial = np.max(exponencial)
min_uniforme = np.min(uniforme)
max_uniforme = np.max(uniforme)
min_poisson = np.min(poisson)
max_poisson = np.max(poisson)
distribuicoes["Normal"]["estatisticas"]["min"] = min_normal
distribuicoes["Normal"]["estatisticas"]["max"] = max_normal
distribuicoes["Exponencial"]["estatisticas"]["min"] = min_exponencial
distribuicoes["Exponencial"]["estatisticas"]["max"] = max_exponencial
distribuicoes["Uniforme"]["estatisticas"]["min"] = min_uniforme
distribuicoes["Uniforme"]["estatisticas"]["max"] = max_uniforme
distribuicoes["Poisson"]["estatisticas"]["min"] = min_poisson
distribuicoes["Poisson"]["estatisticas"]["max"] = max_poisson
df = pd.DataFrame(distribuicoes)
df
```

Passo 9 - Comparar distribuições graficamente

```python
plt.figure(figsize=(10, 6))
plt.boxplot([normal, exponencial, uniforme, poisson], labels=['Normal', 'Exponencial', 'Uniforme', 'Poisson'])
plt.title('Comparação de Distribuições')
plt.ylabel('Valores')
plt.xlabel('Distribuições')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

Passo 10 - Comparar estatísticas calculadas com valores teóricos

```python
teorico_normal = {
    "media": 0,
    "variancia": 1
}


teorico_exponencial = {
    "media": 2,
    "variancia": 4
}

teorico_uniforme = {
    "media": (-1 + 1) / 2,
    "variancia": (1 - (-1))**2 / 12
}


teorico_poisson = {
    "media": 2,
    "variancia": 2
}

comparacao = pd.DataFrame({
    "Distribuição": ["Normal", "Exponencial", "Uniforme", "Poisson"],
    "Média Teórica": [teorico_normal["media"], teorico_exponencial["media"], teorico_uniforme["media"], teorico_poisson["media"]],
    "Média Amostra": [media_normal, media_exponencial, media_uniforme, media_poisson],
    "Variância Teórica": [teorico_normal["variancia"], teorico_exponencial["variancia"], teorico_uniforme["variancia"], teorico_poisson["variancia"]],
    "Variância Amostra": [var_normal, var_exponencial, var_uniforme, var_poisson]
})

print(comparacao)

plt.figure(figsize=(10, 6))
x = np.arange(len(comparacao))
width = 0.35

plt.bar(x - width/2, comparacao["Média Teórica"], width, label="Teórica", alpha=0.8)
plt.bar(x + width/2, comparacao["Média Amostra"], width, label="Amostra", alpha=0.8)

plt.xlabel("Distribuições")
plt.ylabel("Média")
plt.title("Comparação: Média Teórica vs Média Amostra")
plt.xticks(x, comparacao["Distribuição"])
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()
```
